# Basic Analysis

__How many planes have more than 100 seats?__

```python
pd.read_sql_query("""
    select aircraft_code, count(*) as number_of_seats
    from seats
    group by aircraft_code having number_of_seats > 100""", connection)
```

__How the number of tickets booked and total amount earned changed with the time__

```python
tickets = pd.read_sql_query("""
    select * from tickets
    inner join bookings
    on tickets.book_ref = bookings.book_ref""", connection)

tickets['book_date'] = pd.to_datetime(tickets['book_date'])
tickets['date'] = tickets['book_date'].dt.date
```

```python
x = tickets.groupby('date')[['date']].count()
plt.figure(figsize = (18, 6))
plt.plot(x.index, x['date'], marker = '*', color = 'maroon')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Number of tickets', fontsize = 20)
plt.title('Total number of tickets booked around the time', fontsize = 30)
plt.grid()
plt.savefig('Number of tickets vs Date.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
bookings = pd.read_sql_query("""
    select * from bookings""", connection)
bookings['book_date'] = pd.to_datetime(bookings['book_date'])
bookings['date'] = tickets['book_date'].dt.date
```

```python
x = bookings.groupby('date')[['total_amount']].sum()
plt.figure(figsize = (18, 6))
plt.plot(x.index, x['total_amount'], marker = '*', color = 'maroon')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Total Amount', fontsize = 20)
plt.title('Total Amount earned around the time', fontsize = 30)
plt.grid()
plt.savefig('Total amount earned around the time.png', dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

# Analyzing Occupancy Rate

__For each aircraft, calculate the total revenue per year and the average revenue per ticket__

```python
pd.read_sql_query("""
    select aircraft_code, ticket_count, total_revenue, 
    total_revenue/ticket_count as average_revenue_per_ticket from 
    (select aircraft_code, count(*) as ticket_count, sum(amount) as total_revenue
    from ticket_flights join flights on ticket_flights.flight_id = flights.flight_id
    group by aircraft_code)""", connection)
```

__Calculate the average occupancy per aircraft__

```python
occupancy_rate = pd.read_sql_query("""
    select a.aircraft_code, avg(a.seats_count) as booked_seats, b.num_seats, avg(a.seats_count)/b.num_seats as occupancy_rate from
    (select aircraft_code, flights.flight_id, count(*) as seats_count from boarding_passes
    inner join flights on boarding_passes.flight_id = flights.flight_id
    group by aircraft_code, flights.flight_id) as a
    inner join (select aircraft_code, count(*) as num_seats from seats
    group by aircraft_code) as b on a.aircraft_code = b.aircraft_code 
    group by a.aircraft_code""", connection)
occupancy_rate
```

__Calculate by how much the total annual turnover could increase by giving all aircraft a 10% higher occupancy rate__

```python
occupancy_rate['Inc occupancy rate'] = occupancy_rate['occupancy_rate'] + occupancy_rate['occupancy_rate']*0.1
occupancy_rate
```
